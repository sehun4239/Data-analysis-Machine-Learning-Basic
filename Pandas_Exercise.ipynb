```python
## Excercise -1 

import numpy as np
import pandas as pd
import time
from datetime import datetime

df_movie = pd.read_csv('./data/movie/movies.csv')
df_rating = pd.read_csv('./data/movie/ratings.csv')

# display(df_movie)
# display(df_rating)

#1. 사용자가 평가한 모든 영화의 전체 평균 평점
print(df_rating['rating'].mean())



#2. 각 사용자별 평균 평점
df_rating_user = df_rating['rating'].groupby(df_rating['userId']).mean()
print(df_rating_user)



#3. 각 영화별 평균 평점
df_rating_addmovie = pd.merge(df_rating, df_movie, on='movieId', how='inner')
# display(df_rating_addmovie)

df_rating_movie = df_rating_addmovie['rating'].groupby(df_rating_addmovie['movieId']).mean()
df_moviename = pd.merge(df_rating_movie, df_movie, on='movieId', how='inner')
display(df_moviename[['title', 'rating']])


#4. 평균 평점이 가장 높은 영화의 제목(동률이 있을 경우 모두 출력)
print(df_rating_movie[df_rating_movie == df_rating_movie.max()])


#5. Comedy영화 중 가장 평점이 낮은 영화의 제목
df_rating_comedy = df_rating_addmovie.loc[df_rating_addmovie['genres'].str.contains('Comedy')]
# display(df_rating_comedy)

df_low_comedy = df_rating_comedy['rating'].groupby(df_rating_comedy['title']).mean()
print(df_low_comedy[df_low_comedy == df_low_comedy.min()])



#6. 2015년도에 평가된 모든 Romance 영화의 2015년도 평균 평점은?
start_date = datetime(2015,1,1).timestamp()
end_date = datetime(2016,1,1).timestamp()
# print(start_date)
# print(end_date)

df_romance = df_rating_addmovie.loc[df_rating_addmovie['genres'].str.contains('Romance')]
df_2015 = df_romance.loc[(df_romance['timestamp'] >= start_date) & (df_romance['timestamp'] < end_date)]
print(df_2015['rating'].mean())
# print(df_2015['rating'].count())
```

```python
# Exercise 2

import numpy as np
import pandas as pd

df_car = pd.read_csv('./data/mpg.txt')

display(df_car)


# 1. displ(배기량)이 4 이하인 자동차와 5 이상인 자동차 중 
# 어떤 자동차의 hwy(고속도로 연비)가 평균적으로 더 높은지 확인하세요.
df_displ4 = df_car.loc[df_car['displ'] <= 4]
df_displ5 = df_car.loc[df_car['displ'] >= 5]

print('4 이하인 자동차는 {}'.format(df_displ4['hwy'].mean()))
print('5 이상인 자동차는 {}'.format(df_displ5['hwy'].mean()))


# 2. 자동차 제조 회사에 따라 도시 연비가 다른지 알아보려고 한다. 
# "audi"와 "toyota" 중 어느 manufacturer(제조회사)의 cty(도시 연비)가 
# 평균적으로 더 높은지 확인하세요.

df_audi = df_car.loc[df_car['manufacturer'] == 'audi']
df_toyota = df_car.loc[df_car['manufacturer'] == 'toyota']

print('audi cty는 {}'.format(df_audi['cty'].mean()))
print('toyota cty는 {}'.format(df_toyota['cty'].mean()))

# 3. "chevrolet", "ford", "honda" 자동차의 고속도로 연비 평균을 알아보려고 한다. 
# 이 회사들의 데이터를 추출한 후 hwy(고속도로 연비) 평균을 구하세요.
df_cfh = df_car.loc[(df_car['manufacturer'] == 'chevrolet') | (df_car['manufacturer'] == 'ford') | (df_car['manufacturer'] == 'honda')]
print('세 회사 자동차의 hwy는 {}'.format(df_cfh['hwy'].mean()))


# 4. "audi"에서 생산한 자동차 중에 어떤 자동차 모델의 hwy(고속도로 연비)가 
# 높은지 알아보려고 한다. "audi"에서 생산한 자동차 중 hwy가 1~5위에 해당하는 
# 자동차의 데이터를 출력하세요.
display(df_audi.sort_values(by='hwy', ascending=False)[0:5])


# 5. mpg 데이터는 연비를 나타내는 변수가 2개입니다. 
# 두 변수를 각각 활용하는 대신 하나의 통합 연비 변수를 만들어 사용하려 합니다. 
# 평균 연비 변수는 두 연비(고속도로와 도시)의 평균을 이용합니다. 
# 회사별로 "suv" 자동차의 평균 연비를 구한후 내림차순으로 정렬한 후 1~5위까지 데이터를 출력하세요.
df_car['avg'] = (df_car['cty']+df_car['hwy'])/2

df_suv = df_car.loc[df_car['class'] == 'suv']

result5 = df_suv['avg'].groupby([df_suv['manufacturer']]).mean().sort_values(ascending=False)
print(result5[0:5])

# 6. mpg 데이터의 class는 "suv", "compact" 등 자동차의 특징에 따라 
# 일곱 종류로 분류한 변수입니다. 어떤 차종의 도시 연비가 높은지 비교하려 합니다. 
# class별 cty 평균을 구하고 cty 평균이 높은 순으로 정렬해 출력하세요.

result6 = df_car['cty'].groupby([df_car['class']]).mean().sort_values(ascending=False)
print(result6)

# 7. 어떤 회사 자동차의 hwy(고속도로 연비)가 가장 높은지 알아보려 합니다. 
# hwy(고속도로 연비) 평균이 가장 높은 회사 세 곳을 출력하세요.

result7 = df_car['hwy'].groupby([df_car['manufacturer']]).mean().sort_values(ascending=False)
print(result7[0:3])

# 8. 어떤 회사에서 "compact" 차종을 가장 많이 생산하는지 알아보려고 합니다. 
# 각 회사별 "compact" 차종 수를 내림차순으로 정렬해 출력하세요.

df_compact = df_car.loc[df_car['class'] == 'compact']
result8 = df_compact.groupby(df_compact['manufacturer'])
print(result8.size().sort_values(ascending=False))
```

```python
# exercise 3 풀이

# 1. 성별에 따른 월급 차이
# 과거에 비해 여성의 사회 진출이 활발하지만 직장에서의
# 위상에서는 여전히 차별이 존재하고 있는것이 사실.
# 실제로 그러한지 월급의 차이를 이용하여 사실을 확인해보자
print(df['p1002_8aq1'].groupby([df['h10_g3']]).mean())


# 2. 나이와 월급의 관계
# 몇 살 때 월급을 가장 많이 받을까? 또 그때의 월급은 얼마인가?

df_don = df['p1002_8aq1'].groupby([df['h10_g4']]).mean()
df_don.dropna(how='any', inplace=True)
df_don = pd.DataFrame(df_don)
df_don = df_don.sort_values(by='p1002_8aq1', ascending=False).copy()
df_don['age'] = 2015-df_don.index.values

print('{}살 때 월급을 {}만원으로 가장 많이 받는다'.format(df_don.iloc[0,1],df_don.iloc[0,0]))  # 2015년 조사 기준으로

# 3. 연령대에 따른 월급 차이
# 30세 미만을 초년(young), 
# 30~59세 : 중년(middle), 
# 60세 이상 : 노년(old)
# 위의 범주로 연령대에 따른 월급의 차이를 알아보자


def age(x):
    if x['age'] < 30:
        return 'young'
    elif x['age'] >= 60:
        return 'old'
    else:
        return 'middle'

df_don['period'] = df_don.apply(age, axis = 1)
# display(df_don)

print(df_don['p1002_8aq1'].groupby(df_don['period']).mean())

# 4. 연령대 및 성별 월급 차이
# 성별 월급 차이는 연령대에 따라 다른 양상을 보일 수 있습니다.
# 성별 월급 차이가 연령대에 따라 다른지 분석해보자

# 기존에는 3그룹(초년,중년,노년)이었지만 이젠 6그룹으로
# 그룹핑을 해야 한다.(초년남성,초년여성,..)
df['age'] = 2015-df['h10_g4']
df['period'] = df.apply(age, axis = 1)

print(df['p1002_8aq1'].groupby([df['period'],df['h10_g3']]).mean())

# 5. 직업별 월급 차이
# 어떤 직업이 월급을 가장 많이 받을까?
# 직업별 월급을 분석해 보자
# 직업코드는 제공된 Koweps_Codebook.xlsx을 이용하면 
# 편하게 코드값을 이용 할 수 있습니다.

df_jobcode = pd.read_excel('./data/Koweps_Codebook.xlsx',
                           sheet_name = '직종 코드',
                           header = 0)
df_job = pd.merge(df, df_jobcode, left_on='h10_eco9', right_on='code_job', how='inner')
result35 = df_job['p1002_8aq1'].groupby(df_job['job']).mean().sort_values(ascending=False)
print(result35)
print('{}가 가장 월급을 많이 받는다.'.format(result35.index.values[0]))

# 6. 성별 직업 빈도
# 성별로 어떤 직업이 가장 많을까?
df_mfjob = df_job[['h10_g4','h10_g3','job']]
display(df_mfjob)

print(df_mfjob['h10_g3'].groupby(df_mfjob['job']).size())

# 7. 종교 유무에 따른 이혼율
# 종교가 있는 사람들이 이혼을 덜 할까??
df_jonggyo = df['h10_g4'].groupby(df['h10_g11']).size()
display(df_jonggyo)


# 8. 지역별 연령대 비율
# 노년층이 많은 지역은 어디일까?



```
